```python
import pandas as pd
```

```python
df = pd.read_json('data/twitter_data2020-01-28_12:44.json', lines=True)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H%M%S')
```

```python
daily_likes = df.groupby(df['date'].apply(lambda d: (d.year, d.month, d.day)))['likes'].sum()
```

```python
daily_likes.plot(kind='density')
```
